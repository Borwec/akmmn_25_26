```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

# Перший датасет - Credit Card Fraud Detection

```python
# https://www.kaggle.com/datasets/nelgiriyewithana/credit-card-fraud-detection-dataset-2023

from google.colab import drive
drive.mount("/content/drive")

df = pd.read_csv("/content/drive/MyDrive/akmmn_25_26/lab2/creditcard_2023.csv")
```

```python
X, y = df.iloc[:, 1:-1], df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state = 127)
```

```python
dt = DecisionTreeClassifier(random_state = 127)
dt.fit(X_train, y_train)
pred = dt.predict(X_test)

print(classification_report(y_test,pred))
print()
print(confusion_matrix(y_test,pred))
print()
print(accuracy_score(y_test,pred))
```

```python
rf = RandomForestClassifier(random_state = 127, n_jobs=-1)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)

print(classification_report(y_test,pred))
print()
print(confusion_matrix(y_test,pred))
print()
print(accuracy_score(y_test,pred))
```

# Третій датасет - Web page Phishing Detection

```python
# https://www.kaggle.com/datasets/shashwatwork/web-page-phishing-detection-dataset

url = "https://drive.google.com/uc?id=1GHvx-dcoEcq5ToNqJG_zrwzGI9LZzJEs"
df = pd.read_csv(url)
```

```python
df = df.drop("url", axis=1)
df["status"] = df["status"].map({"legitimate":0, "phishing":1})
```

```python
X, y = df.iloc[:, :-1], df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state = 127)
```

```python
dt = DecisionTreeClassifier(random_state = 127)
dt.fit(X_train, y_train)
pred = dt.predict(X_test)

print(classification_report(y_test,pred))
print()
print(confusion_matrix(y_test,pred))
print()
print(accuracy_score(y_test,pred))
```

```python
rf = RandomForestClassifier(random_state = 127, n_jobs=-1)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)

print(classification_report(y_test,pred))
print()
print(confusion_matrix(y_test,pred))
print()
print(accuracy_score(y_test,pred))
```

# Висновки
Перевагами алгоритму дерев рішень є їх здатність вловлювати нелінійні залежності, можливість працювати з різними типами даних (числовими, категоріальними), відсутність чутливості до масштабу даних, що прибирає потребу в певних кроках попередньої обробки даних, а також стійкість до винятків

Їх слабкими сторонами є схильність до перенавчання, значна нестабільність та неможливість передбачувати нові, не використані при навчанні дані. Частину цих проблем вирішують підбором параметрів, зокрема обмеженням глибини дерева, а також використанням багатьох дерев, як в алгоритмі випадкового лісу, ціною вищої потреби в пам'яті та обчислювальних потужностях

Найефективніше дерева рішень та їх ансамблі показують себе на нелінійних даних, різні типи даних також підвищують доцільність їх використання
