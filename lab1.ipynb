```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# https://www.kaggle.com/datasets/piyushrumao/malware-executable-detection

url = "https://drive.google.com/uc?id=1y1MMqaPDZclUx1emxy37t_MYfAh9CcYo"
df = pd.read_csv(url)
```

```python
X, y = df.iloc[:, 1:], df.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state = 127)
```

```python
logreg = LogisticRegression(random_state=127).fit(X_train, y_train)
pred = logreg.predict(X_test)
```

```python
print(classification_report(y_test,pred))
print()
print(confusion_matrix(y_test,pred))
print()
print(accuracy_score(y_test,pred))
```
