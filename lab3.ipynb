```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics.pairwise import rbf_kernel
```

```python
# https://www.kaggle.com/datasets/shashwatwork/web-page-phishing-detection-dataset

url = "https://drive.google.com/uc?id=1GHvx-dcoEcq5ToNqJG_zrwzGI9LZzJEs"
df = pd.read_csv(url)
```

```python
df = df.drop("url", axis=1)
df["status"] = df["status"].map({"legitimate":0, "phishing":1})
```

```python
X, y = df.iloc[:, :-1], df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state = 127)
```

```python
svm = SVC(kernel=rbf_kernel)
svm.fit(X_train, y_train)
pred = svm.predict(X_test)

print(classification_report(y_test,pred))
print()
print(confusion_matrix(y_test,pred))
print()
print(accuracy_score(y_test,pred))
```

```python
params = {"C": [10 ** x for x in range(-5, 6)], "gamma": [10 ** -x for x in range(0, 6)]}
grid = GridSearchCV(SVC(kernel=rbf_kernel), params, cv=5, n_jobs=-1)

grid.fit(X_train, y_train)
grid.best_params_
```

```python
pred = grid.predict(X_test)

print(classification_report(y_test,pred))
print()
print(confusion_matrix(y_test,pred))
print()
print(accuracy_score(y_test,pred))
```
