```python
import pandas as pd
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# https://www.kaggle.com/datasets/bassamkasasbeh1/wsnds

url = "https://drive.google.com/uc?id=17l8bicDiZyf2HQUlTJNuqZDV-dFcJaX1"
df = pd.read_csv(url)
```

```python
df = df.drop([" id", " who CH", " Time"], axis=1)
```

```python
X = df.iloc[:, :-1]
X = StandardScaler().fit_transform(X)
X = PCA(n_components=2).fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=5, random_state=127)
kmeans.fit(X)
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], marker='o', c='red', s=200, alpha=0.8)
plt.show()
```

```python
x = X[:len(X) // 10]
```

```python
agglo = AgglomerativeClustering(n_clusters=5)
agglo.fit(x)
plt.scatter(x[:, 0], x[:, 1], c=agglo.labels_)
plt.show()
```
