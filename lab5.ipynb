```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# https://www.kaggle.com/datasets/bassamkasasbeh1/wsnds

url = "https://drive.google.com/uc?id=17l8bicDiZyf2HQUlTJNuqZDV-dFcJaX1"
df = pd.read_csv(url)
```

```python
df = df.drop([" id", " who CH", " Time"], axis=1)
```

```python
X, y = df.iloc[:, :-1], df.iloc[:, -1]
X = StandardScaler().fit_transform(X, y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state = 127)
```

```python
knn = KNeighborsClassifier(n_neighbors=3, n_jobs=-1)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)

print(classification_report(y_test,pred))
print()
print(confusion_matrix(y_test,pred))
print()
print(accuracy_score(y_test,pred))
```

# Висновки
Алгоритм K найближчих сусідів є простим в реалізації, гнучким для різних завдань і не потребує попереднього навчання. Його недоліками є чутливість до шуму, висока потреба в обчислювальних ресурсах та чутливість до параметру K кількості найближчих сусідів

Точність алгоритму залежить від того, наскільки залежності даних можна пояснити відстанню. Через те, що KNN базується на відстані, він потребує зведення даних до єдиного масштабу. Також він втрачає ефективність на даних з великою кількістю ознак, адже у вищих вимірах поняття відстані починає втрачати значимість
